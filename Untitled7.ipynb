```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("sample_movies.csv")  # or your own dataset
movies.columns = movies.columns.str.strip()
movies = movies[['title', 'overview']].dropna()
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['overview'])
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend_movie(title):
    # Check if the title exists in the movies DataFrame
    if title not in movies['title'].values:
        print(f"Movie '{title}' not found in the dataset.")
        return {} # Return an empty dictionary if movie not found

    index = movies[movies['title'] == title].index[0]
    distances = similarity[index]
    # Get the indices and similarity scores of the top 5 most similar movies (excluding the input movie itself)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Create a dictionary with the input movie title as the key and the list of recommendations as the value
    recommendations = {title: [movies.iloc[i[0]].title for i in movie_list]}

    return recommendations
```

```python
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt

# Fix: simple parser for stringified genre list
def extract_genres(x):
    try:
        return ast.literal_eval(x)
    except:
        return []

# Load the sample dataset
movies_genre = pd.read_csv("sample_movies_with_genres.csv")

# Extract genres from string
movies_genre['genres_list'] = movies_genre['genres'].apply(extract_genres)
movies_genre['genre_str'] = movies_genre['genres_list'].apply(lambda x: ', '.join(x))

# Count top genres
all_genres = sum(movies_genre['genres_list'].tolist(), [])
genre_counts = Counter(all_genres)

# Plot
plt.figure(figsize=(10, 5))
plt.bar(*zip(*genre_counts.most_common(10)), color='skyblue')
plt.title("Top 10 Movie Genres")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
print("Plotting genre count now.")

```

```python
import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
movies = pd.read_csv("sample_movies.csv")  # Use your own dataset

# Fill missing overviews
movies['overview'] = movies['overview'].fillna("")

# Vectorize overviews
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['overview'])
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Index for title lookup
movies = movies.reset_index()
indices = pd.Series(movies.index, index=movies['title'].str.lower())

# Recommendation function
def recommend(title, n=5):
    title = title.lower()
    if title not in indices:
        return []
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n+1]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]

# Streamlit interface
st.title(" Movie Recommendation System")

movie_input = st.text_input("Enter a movie title:")

if movie_input:
    results = recommend(movie_input, n=5)
    if results:
        st.subheader("Recommended Movies:")
        for movie in results:
            st.write(f"- {movie}")
    else:
        st.error("Movie not found in the dataset.")
```
